```python
import pandas as pd
from statsmodels.tsa.arima_model import  ARIMA,ARMA
import matplotlib.pyplot as plt
import calendar
from datetime import datetime, timedelta
from itertools import product
import numpy as np
import statsmodels.api as sm
import warnings
warnings.filterwarnings("ignore")
from keras.models import Sequential
from keras.layers import Dense
from keras.layers import LSTM
from keras.layers import Dropout
from pandas import DataFrame
from pandas import concat
from itertools import chain
from sklearn.metrics import mean_squared_error
from sklearn.preprocessing import MinMaxScaler

plt.rcParams['font.sans-serif'] = ['SimHei'] 
```

```python
data1=pd.read_excel("./区房价/北京区房价.xlsx",sheet_name = None)
data2=pd.read_excel("./区房价/上海区房价.xlsx",sheet_name = None)
data3=pd.read_excel("./区房价/广州区房价.xlsx",sheet_name = None)
data4=pd.read_excel("./区房价/深圳区房价.xlsx",sheet_name = None)

BeiJing={}
ShangHai={}
GuangZhou={}
ShenZheng={}
for area,df in data1.items():
    BeiJing[area]=df
for area,df in data2.items():
    ShangHai[area]=df
for area,df in data3.items():
    GuangZhou[area]=df
for area,df in data4.items():
    ShenZheng[area]=df
     

```

```python
def predicthouseprice(city,area,df):

    df['时间(年-月)']=pd.to_datetime(df['时间(年-月)'])
    df.sort_values('时间(年-月)',ascending=True,inplace=True)
    tmp=df.loc[:,['时间(年-月)','供给(元/㎡)']].set_index(['时间(年-月)'])
    # 设置参数范围
    ps = range(0, 7)
    qs = range(0, 7)

    parameters = product(ps,qs)
    parameters_list = list(parameters)
    # 寻找最优ARMA模型参数，即best_aic最小
    results = []
    best_aic = float("inf") # 正无穷
    for param in parameters_list:
        try:
            model  =ARMA(tmp,order=(param[0], param[1])).fit()

#             model = ARIMA(tmp,order=(param[0], param[1], param[2])).fit()

        except :
            print('参数错误:', param)
            continue
     
        aic = model.aic
        if aic < best_aic:
            best_model = model
            best_aic = aic
            best_param = param
        results.append([param, model.aic])
    print('最优模型: ', best_model.summary())

    # 设置future_month，需要预测的时间date_list
    future_month = 3
    last_month = pd.to_datetime(tmp.index[len(tmp)-1])
    date_list = []
    for i in range(future_month):
        # 计算这个月有多少天
        year = last_month.year
        month = last_month.month

        this_month_days = calendar.monthrange(year, month)[1]
        #print(next_month_days)
        last_month = last_month + timedelta(days=this_month_days)
        date_list.append(last_month)
    print('date_list=', date_list)


    print(tmp.columns)
    # 添加未来要预测的3个月
    future = pd.DataFrame(index=date_list, columns= tmp.columns)
    tmp = pd.concat([tmp, future])
#    tmp['forecast'] = best_model.get_prediction(start=0, end=len(tmp)).predicted_mean 
    tmp['forecast'] = best_model.predict(start=0, end=len(tmp)) 
    tmp['forecast'][0] = tmp['供给(元/㎡)'][0]
    print(tmp)
    tmp.to_csv('%s,%s.csv'%(city,area))
    # 沪市指数预测结果显示
    plt.figure(figsize=(30,7))
    tmp['供给(元/㎡)'].plot(label='实际指数')
    tmp.forecast.plot(color='r', ls='--', label='预测指数')
    plt.legend()
    plt.title('房价')
    plt.xlabel('时间')
    plt.ylabel('房价')
    plt.show()




```

```python
# 转化为可以用于监督学习的数据
def get_train_set(data_set, timesteps_in, timesteps_out=1):
    train_data_set = np.array(data_set)
    reframed_train_data_set = np.array(series_to_supervised(train_data_set, timesteps_in, timesteps_out).values)
#    print(reframed_train_data_set)
#    print(reframed_train_data_set.shape)
    train_x, train_y = reframed_train_data_set[:, :-timesteps_out], reframed_train_data_set[:, -timesteps_out:]
    # 将数据集重构为符合LSTM要求的数据格式,即 [样本数，时间步，特征]
    train_x = train_x.reshape((train_x.shape[0], timesteps_in, 1))
    return train_x, train_y
def series_to_supervised(data, n_in=1, n_out=1, dropnan=True):
    n_vars = 1 if type(data) is list else data.shape[1]
    df = DataFrame(data)
    cols, names = list(), list()
    # input sequence (t-n, ... t-1)
    for i in range(n_in, 0, -1):
        cols.append(df.shift(i))
        names += [('var%d(t-%d)' % (j + 1, i)) for j in range(n_vars)]
    # 预测序列 (t, t+1, ... t+n)
    for i in range(0, n_out):
        cols.append(df.shift(-i))
        if i == 0:
            names += [('var%d(t)' % (j + 1)) for j in range(n_vars)]
        else:
            names += [('var%d(t+%d)' % (j + 1, i)) for j in range(n_vars)]
    # 拼接到一起
    agg = concat(cols, axis=1)
    agg.columns = names
    # 去掉NaN行
    if dropnan:
        agg.dropna(inplace=True)
    return agg
def lstm_model(city,area,source_data_set, df,train_x, label_y, input_epochs, input_batch_size, timesteps_out):

#    train_x = preprocessing.scale(train_x)
    model = Sequential()
    
    # 第一层, 隐藏层神经元节点个数为128, 返回整个序列
    model.add(LSTM(128, return_sequences=True, activation='tanh', input_shape=(train_x.shape[1], train_x.shape[2])))
    # 第二层，隐藏层神经元节点个数为128, 只返回序列最后一个输出
    model.add(LSTM(128, return_sequences=False))
    model.add(Dropout(0.5))
    # 第三层 因为是回归问题所以使用linear
    model.add(Dense(timesteps_out, activation='linear'))
    model.compile(loss='mean_squared_error', optimizer='adam')

    # LSTM训练 input_epochs次数
    res = model.fit(train_x, label_y, epochs=input_epochs, batch_size=input_batch_size, verbose=0, shuffle=False)

    # 模型预测
    train_predict = model.predict(train_x)
    predict=[scaler.inverse_transform(x[0].reshape(1,-1))[0][0] for x in  train_predict]
    tmp=df  
    predict_tail=[scaler.inverse_transform(x.reshape(1,-1))[0][0] for x in  train_predict[-1]][-2:]
     

    print(label_y[-1])
    
    train_x_tail = np.array(label_y[-1]).reshape((1, timesteps_in, 1))
    predict_last=model.predict(train_x_tail)
    predict_last=[scaler.inverse_transform(x.reshape(1,-1))[0][0] for x in  predict_last[0]]
    
   
    plt.plot(res.history['loss'], label='train')
    plt.show()
    print(model.summary())

    plt.figure(figsize=(24, 8))
    # 原始数据蓝色

    
# 设置future_month，需要预测的时间date_list
    future_month = 3
    last_month = pd.to_datetime(tmp.index[len(tmp)-1])
    date_list = []
    for i in range(future_month):
        # 计算这个月有多少天
        year = last_month.year
        month = last_month.month

        this_month_days = calendar.monthrange(year, month)[1]
        #print(next_month_days)
        last_month = last_month + timedelta(days=this_month_days)
        date_list.append(last_month)
#    print('date_list=', date_list)


 #   print(tmp.columns)
    # 添加未来要预测的3个月
    future = pd.DataFrame(index=date_list, columns= tmp.columns)
    tmp = pd.concat([tmp, future])
#    tmp['forecast'] = best_model.get_prediction(start=0, end=len(tmp)).predicted_mean 
 #   print(len(tmp),len(predict),len(predict_tail),len(predict_last)) 
#预测列的前三个数直接用原真实值
    tmp['forecast'] =list(tmp['供给(元/㎡)'][:3])+predict+predict_tail+predict_last

    print(tmp)
    tmp.to_csv('%s,%s.csv'%(city,area))

    plt.figure(figsize=(30,7))
    tmp['供给(元/㎡)'].plot(label='实际指数')
    tmp.forecast.plot(color='r', ls='--', label='预测指数')
    plt.legend()
    plt.title('房价')
    plt.xlabel('时间')
    plt.ylabel('房价')
    plt.show()
    
    
```

```python
def predicthousepricewithlstm(city,area,df):
    df['时间(年-月)']=pd.to_datetime(df['时间(年-月)'])
    df.sort_values('时间(年-月)',ascending=True,inplace=True)
    tmp=df.loc[:,['时间(年-月)','供给(元/㎡)']].set_index(['时间(年-月)'])
    #tmp=BeiJing['昌平区'].loc[:,['供给(元/㎡)']]
    timesteps_in = 3
    timesteps_out = 3
    epochs = 500
    batch_size = 100
    #tmp = preprocessing.scale(tmp)

    scaler = MinMaxScaler(feature_range=(0, 1))
    tmp0 = scaler.fit_transform(tmp)

    train_x, label_y = get_train_set(tmp0, timesteps_in=timesteps_in, timesteps_out=timesteps_out)
    lstm_model(city,area,tmp0,tmp, train_x, label_y, epochs, batch_size, timesteps_out=timesteps_out)
```
